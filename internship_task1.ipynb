```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
df

pd.DataFrame(df)

dataset = df.iloc[1:16,:]
dataset
```

```python
dataset.info
```

```python
plt.figure(figsize=(10, 6))
plt.bar(dataset['title'], dataset['imdbAverageRating'], color='#2e86c1')
plt.xticks(rotation=30, ha='right')
plt.xlabel('Movie Name')
plt.ylabel('IMDb Average Rating')
plt.title('IMDb Average Ratings of First 15 Titles')
plt.tight_layout()
plt.show()
```
